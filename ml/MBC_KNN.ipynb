```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
import random as rd
import pickle


df = pd.read_csv('dataml1.csv')
df
```

```python
asisten = df.drop(columns=['Nama Panggil', 'Jenis Kelamin', 'Jurusan', 'Angkatan', 'Divisi', 'Status', 'Banyak Project'])
asisten
```

```python
bj = asisten.groupby(['Skill'])['Name'].sum().reset_index()
bj
```

```python
bj.str.lower()
```

```python
from sklearn import preprocessing
  
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

asisten['Skill'] = asisten['Skill'].str.lower()

# Encode labels in column 'species'.
asisten['Divisi']= label_encoder.fit_transform(asisten['Divisi']) 
asisten['Divisi'].unique()

asisten['Status']= label_encoder.fit_transform(asisten['Status']) 
asisten['Status'].unique()

asisten['Jurusan']= label_encoder.fit_transform(asisten['Jurusan']) 
asisten['Jurusan'].unique()

asisten['Skills']= label_encoder.fit_transform(asisten['Skill']) 
asisten['Skills'].unique()

feature = ['Angkatan', 'Divisi', 'Status', 'Skills']

asisten
```

```python
list_skill = asisten['Skill'].str.strip('[]').str.replace(' ','').str.replace("'",'')
list_skill = list_skill.str.split(',')
```

```python

x = asisten[feature].copy()
y = asisten['Name']

knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(x,y)
```

```python
new_asisten = asisten
new_asisten['Pred_Name'] = knn.predict(new_asisten[feature])
new_asisten
```

```python
nama_asisten = new_asisten['Pred_Name']

def input_skill(inputan):
    for i in range(len(new_asisten['Skill'])):
        for j in range(len(list_skill[i])):
            if (inputan == list_skill[i][j]):
                z = [] 
                z.append(i)
                print(z)
                print(nama_asisten.iloc[z])
    return(nama_asisten.iloc[z])
```
